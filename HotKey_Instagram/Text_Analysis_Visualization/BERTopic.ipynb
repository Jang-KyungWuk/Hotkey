```python
from tqdm import tqdm
from sklearn.feature_extraction.text import CountVectorizer
from kiwipiepy import Kiwi
from bertopic import BERTopic
import pickle as pkl
```

```python
kiwi = Kiwi()
```

```python
with open("C:\\Users\\user\\recent_list.pkl","rb") as file:
     data = pkl.load(file)
```

```python
print(data)
```

```python
# documents = [line.strip() for line in data[0]]
documents = [data[n] for n in range(500)]
```

```python
preprocessed_documents = []

for line in tqdm(documents):
  # 빈 문자열이거나 숫자로만 이루어진 줄은 제외
  if line and not line.replace(' ', '').isdecimal():
    preprocessed_documents.append(line)
```

```python
callCount=0
```

```python
class CustomTokenizer:
    def __init__(self, tagger):
        self.tagger = tagger
    def __call__(self, sent):
        sent = sent[:1000000]
        
        word_tokens = list()
        for tok in self.tagger.tokenize(sent):
            word_tokens.append(tok.form)
        result = [word for word in word_tokens if len(word) > 1]
        
        global callCount
        callCount+=1
        
        return result
```

```python
custom_tokenizer = CustomTokenizer(Kiwi())
```

```python
vectorizer = CountVectorizer(tokenizer=custom_tokenizer, max_features=5)
```

```python
model = BERTopic(embedding_model="sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens", \
                 vectorizer_model=vectorizer,
                 nr_topics=3,
                 top_n_words=6,
                 calculate_probabilities=True)
```

```python
topics, probs = model.fit_transform(preprocessed_documents) #모델 만드는것
```
