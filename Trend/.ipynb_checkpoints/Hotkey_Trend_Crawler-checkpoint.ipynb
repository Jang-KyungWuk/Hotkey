```python
#한국데이터거래소 100개, 구글 트렌드 인기검색어 전부
from requests import get, Session
import json
params = dict()
trendlist = list()
session = Session()
url1 = 'https://dable-public.s3-ap-northeast-1.amazonaws.com/static/production/tmp/media-index/search_word.json?v=1669045216030'
url2 = 'https://trends.google.co.kr/trends/api/dailytrends?hl=ko&tz=-540&geo=KR&ns=15'
resp1 = session.get(url1)
resp2 = session.get(url2)
for i in range(100):
    trendlist.append(resp1.json()['data'][i]['keyword'])

while len(json.loads(resp2.text[6:])['default']['trendingSearchesDays']):
    params['ed']=json.loads(resp2.text[6:])['default']['endDateForNextRequest']
    resp2 = session.get(url2, params = params)
    for i in range(len(json.loads(resp2.text[6:])['default']['trendingSearchesDays'])):
        for j in range(len(json.loads(resp2.text[6:])['default']['trendingSearchesDays'][i]['trendingSearches'])):
            trendlist.append(json.loads(resp2.text[6:])['default']['trendingSearchesDays'][i]['trendingSearches'][j]['title']['query'])

#trendlist 정제 (띄어쓰기 예외처리)
results = list()
for word in trendlist:
    split = word.split()
    if len(split) > 1:
        results.append(word.replace(" ", "")) #붙여버리기
        for i in split:
            results.append(i) #각각 따로 넣기
    elif len(split) == 1:
        results.append(word)

#중복제거
results = list(set(results))
len(results), results
```

```python
def Trend_Crawler():
    #한국데이터거래소 100개, 구글 트렌드 인기검색어 최신일자
    from requests import get, Session
    import json
    trendlist = list()
    session = Session()
    url1 = 'https://dable-public.s3-ap-northeast-1.amazonaws.com/static/production/tmp/media-index/search_word.json?v=1669045216030'
    url2 = 'https://trends.google.co.kr/trends/api/dailytrends?hl=ko&tz=-540&geo=KR&ns=15'
    resp1 = session.get(url1)
    resp2 = session.get(url2)
    for i in range(100):
        trendlist.append(resp1.json()['data'][i]['keyword'])

    for i in range(len(json.loads(resp2.text[6:])['default']['trendingSearchesDays'][0]['trendingSearches'])):
        trendlist.append(json.loads(resp2.text[6:])['default']['trendingSearchesDays'][0]['trendingSearches'][i]['title']['query'])

    #trendlist 정제 (띄어쓰기 예외처리)
    results = list()
    for word in trendlist:
        split = word.split()
        if len(split) > 1:
            results.append(word.replace(" ", "")) #붙여버리기
            for i in split:
                results.append(i) #각각 따로 넣기
        elif len(split) == 1:
            results.append(word)

    #중복제거
    results = list(set(results))
    return results
```
