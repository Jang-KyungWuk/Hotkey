```python
from requests import get, post, Session
from urllib import parse
import json
from datetime import datetime
import pickle as pkl
from konlpy.tag import Okt
okt = Okt()
```

```python
with open('../../recent_list.pkl', 'rb') as f:
    b = pkl.load(f)
```

```python
oktnounlist=list()
for i in b:
    oktnounlist=oktnounlist+okt.nouns(i)
```

```python
oktnounlist
```

```python
voca = dict()
for term in oktnounlist:
    if term in voca:
        voca[term] += 1
    else:
        voca[term] = 1
    
```

```python
voca
```

```python
sorted_voca = sorted(voca.items(), key = lambda item: item[1], reverse = True)
print(sorted_voca)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
font_path ='C:\Windows\Fonts\malgun.ttf'
# fontprop = fm.FontProperties(fname=font_path, size=18)
im = Image.open('../../Twitter.svg') # 이미지 파일 읽어오기
mask_arr = np.array(im)
wc=WordCloud(background_color='white',font_path = font_path, mask = mask_arr).generate_from_frequencies(voca)
```

```python
plt.figure(figsize=(15,8))
plt.imshow(wc)
```

```python
wordcloud = WordCloud(background_color ='white', colormap='autumn',
                      width = 700, height = 700, font_path = font_path,random_state = 43, mask = mask_arr,
                      prefer_horizontal = True).generate_from_frequencies(voca)

plt.figure(figsize = (6, 6))  
plt.imshow(wordcloud)
plt.title("Word Frequency", size = 13)

plt.axis('off')

plt.show()
```
