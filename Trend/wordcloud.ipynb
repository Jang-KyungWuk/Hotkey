# Okt의 noun으로 토큰 생성

```python
import pickle as pkl
from konlpy.tag import Okt
okt = Okt()
with open('../../recent_list.pkl', 'rb') as f:
    b = pkl.load(f)
oktnounlist=list()
for i in b:
    oktnounlist=oktnounlist+okt.nouns(i)
voca = dict()
for term in oktnounlist:
    if term in voca:
        voca[term] += 1
    else:
        voca[term] = 1
```

# word cloud 생성

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
font_path ='C:\Windows\Fonts\malgun.ttf'
# fontprop = fm.FontProperties(fname=font_path, size=18)
im = Image.open('../../twitter_mask2.png') # 이미지 파일 읽어오기
mask=Image.new("RGB",im.size, (255,255,255))
mask.paste(im)
mask=np.array(mask)
wc=WordCloud(background_color='white',colormap='winter', font_path = font_path, mask = mask)
wc=wc.generate_from_frequencies(voca)
plt.imshow(wc)
```

# Kiwi에서 쓸만한 tag 확인

```python
kiwitokenlist=[]
for i in b:
    for j in kiwi.tokenize(i):
        if j.tag in tagset:
            kiwitokenlist.append(j.form)
```

```python
voca = dict()
for term in kiwitokenlist:
    if term in voca:
        voca[term] += 1
    else:
        voca[term] = 1
```

```python
from kiwipiepy import Kiwi
kiwi = Kiwi()
kiwi.prepare()
alltagset = set()
for i in b:
    for j in kiwi.tokenize(i):
        alltagset.add(j.tag)
for k in alltagset:
    print(k, '\n')
    for i in b:
        for j in kiwi.tokenize(i):
            if j.tag == k:
                print(j.form, end=',')
    print('-------------------------------------\n')
```
