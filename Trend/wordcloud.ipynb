```python
from requests import get, post, Session
from urllib import parse
import json
from datetime import datetime
import pickle as pkl
from konlpy.tag import Okt
okt = Okt()
```

```python
with open('../../recent_list.pkl', 'rb') as f:
    b = pkl.load(f)
```

```python
oktnounlist=list()
for i in b:
    oktnounlist=oktnounlist+okt.nouns(i)
```

```python
oktnounlist
```

```python
voca = dict()
for term in oktnounlist:
    if term in voca:
        voca[term] += 1
    else:
        voca[term] = 1
    
```

```python
voca
```

```python
sorted_voca = sorted(voca.items(), key = lambda item: item[1], reverse = True)
print(sorted_voca)
```
