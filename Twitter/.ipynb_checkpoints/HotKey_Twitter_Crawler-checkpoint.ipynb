```python
def hot_key_twitter(keyword, nposts): #keyword는 검색할 변수, nposts는 크롤링하고자하는 게시물 수
    #반환할 textlist
    textlist = list()
    ncrawled = 0
    
    #parameter 초기값 받아오기
    params, headers = set_params(keyword)
    
    #session 인스턴스 생성
    session = Session()
    request_url = 'https://twitter.com/i/api/2/search/adaptive.json'
    
    #get 요청을 보낼 url 완성
    keys = list(params.keys())
    values = list(params.values())
    get_request_url = request_url+'?'
    for cursor in range(len(keys)):
        get_request_url+='%s=%s&'%(keys[cursor],values[cursor]) 
    get_request_url = get_request_url[:-1]
    
    #headers의 'path' 변수 변경
    headers['path'] = get_request_url[19:]
    
    #get 요청 보내기 (1페이지)
    resp = session.get(get_request_url,headers=headers)
    
    #1페이지 검색
    print('---------------------------------------수집중 : {}페이지...-----------------------------------------\n\n'.format(1))
    textlist.append(list())
    print(resp.json()['globalObjects']['tweets'].keys())
    for i in resp.json()['globalObjects']['tweets'].keys():
        print(resp.json()['globalObjects']['tweets'][i]['full_text'])
        textlist[0].append(resp.json()['globalObjects']['tweets'][i]['full_text'])
        if ncrawled > nposts:
            print('수집 완료!')
            return textlist
        ncrawled+=1
        
    #다음 페이지 가르키는 커서 변수에 값 넣어주기
    next_pg = resp.json()['timeline']['instructions'][0]['addEntries']['entries'][-1]['content']['operation']['cursor']['value']
    
    #2페이지부터 추가 검색
    cur_page = 2
    while True:
        params['cursor'] = next_pg
        get_request_url = request_url+'?'
        keys = list(params.keys())
        values = list(params.values())
        for cursor in range(len(keys)):
            get_request_url+='%s=%s&'%(keys[cursor],values[cursor]) 
        get_request_url = get_request_url[:-1]
        resp = session.get(get_request_url,headers=headers)
        next_pg = resp.json()['timeline']['instructions'][-1]['replaceEntry']['entry']['content']['operation']['cursor']['value']
        print('\n\n---------------------------------------수집중 : {}페이지...-----------------------------------------\n\n'.format(cur_page))
        textlist.append(list())
        print(resp.json()['globalObjects']['tweets'].keys())
        for i in resp.json()['globalObjects']['tweets'].keys():
            print(resp.json()['globalObjects']['tweets'][i]['full_text'])
            textlist[-1].append(resp.json()['globalObjects']['tweets'][i]['full_text'])
            if ncrawled > nposts:
                print('수집 완료!')
                return textlist
            ncrawled+=1
        cur_page+=1
    
    print('수집 완료! 수집한 게시물 수 :', ncrawled)
    return textlist
```
